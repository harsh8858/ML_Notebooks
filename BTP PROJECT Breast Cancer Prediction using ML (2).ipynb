# Breast Cancer Prediction(Malignant or benign) using ML

```python
import numpy as np # library used to make numpy arrays
import pandas as pd # Used to make pandas dataframe
import sklearn.datasets # for getting the data of breast cancer
from sklearn.model_selection import train_test_split # used for training and testing the model
from sklearn.linear_model import LogisticRegression # model that we were using for the prediction
from sklearn.metrics import accuracy_score # used to find the accuracy of the model
import matplotlib.pyplot as plt
import seaborn as sns
```

 Code for Data collecting and processing:

```python
#Getting the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#In above dataset we have various features on which our prediction wil based upon and the target gives us the value whether our cancer is malignant(0) or benign(1)
```

For Loading the above data into the pandas dataframe:

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
```

```python
#In the above code we are getting the values of the features using the 'data' array and marking their names using array 'feature_names'
```

Printing the first five rows of our data frame using 'head'

```python
data_frame.head()
```

```python
#adding the 'target' column to the data frame
data_frame['cancer']= breast_cancer_dataset.target
```

```python
#Again printing the first 5 rows of our data frame
data_frame.head()
```

```python
#printing the last 5 rows of our data frame
data_frame.tail()
```

```python
#Getting the number of examples and features in our data
data_frame.shape
```

```python
#for checking if any feature has null value
data_frame.isnull().sum()
```

```python
# some statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Cancer
data_frame['cancer'].value_counts()
```

```python
# Taking some describe value of each feature for both the cancers
data_frame.groupby('cancer').mean()
```

```python
sns.countplot(data_frame['cancer'], label= 'count')
```

```python
sns.pairplot(data_frame, hue = 'cancer', 
             vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'] )
```

```python
# Seperating the Features and the Target value:



```

```python
X = data_frame.drop(columns='cancer',axis=1) #Giving all the data to X excluding the column cancer
Y = data_frame['cancer'] #Giving the target/cancer value to Y
```

```python
print(X) #Does not contain taget
```

```python
print(Y) #Contain only target
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2) # Giving 80% to training data and 20% to test data
```

```python
#Getting how our original feature data get split into train and test
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING

Logistic Regression

```python
model = LogisticRegression()
```

```python
#Training the logistic regression model using training data

model.fit(X_train, Y_train)
```

Now Model Evaluation Using Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy on the training data = ',training_data_accuracy) #printing the accuracy on training data
```

We can see that Accuracy of our model on Training data is about <b>95%</b>.

```python
#But what matter most is accuracy on our test data
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print('Accuracy on the test data = ',test_data_accuracy) #printing the accuracy on test data
```

We get <b>Accuracy on the test data about 93%</b>.

By seeing the accuracy we can say that our model is not overfitting or underfitting the data
